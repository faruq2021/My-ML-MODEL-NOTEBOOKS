```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/Test.csv")
```

```python
df.head(20)
```

```python
# Check for missing values
print(df.isnull().sum())

```

```python
# Impute missing Age values with the median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df['Time_of_service'].fillna(df['Time_of_service'].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df[['VAR2', 'VAR4']] = df[['VAR2', 'VAR4']].fillna(df[['VAR2', 'VAR4']].mean())
```

```python
print(df.isnull().sum())
```

```python
new_df = df
```

```python
new_df.head()
```

```python
print(new_df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
# Create a LabelEncoder object
le = LabelEncoder()

# Fit and transform the Gender column
new_df['Gender'] = le.fit_transform(new_df['Gender'])

```

```python
new_df.head()
```

```python
new_df['Relationship_Status'] = le.fit_transform(new_df['Relationship_Status'])
```

```python
new_df.head()
```

```python
print(new_df['Hometown'].unique())
print(new_df['Unit'].unique())
```

```python

# Fit and transform the Hometown column
new_df['Hometown'] = le.fit_transform(new_df['Hometown'])

# Create a dictionary to store the mapping
hometown_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

print("Hometown Mapping:")
print(hometown_mapping)

# Fit and transform the Unit column
new_df['Unit'] = le.fit_transform(new_df['Unit'])

# Create a dictionary to store the mapping
unit_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

print("\nUnit Mapping:")
print(unit_mapping)
```

```python
# Fit and transform the Unit column
new_df['Decision_skill_possess'] = le.fit_transform(new_df['Decision_skill_possess'])

# Create a dictionary to store the mapping
Decision_skill_possess_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
print("\nDecision_skill_possess_mapping:")
print(Decision_skill_possess_mapping)
```

```python
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['Age', 'Time_of_service', 'Time_since_promotion', 'VAR1', 'VAR2', 'VAR3', 'VAR4', 'VAR5', 'VAR6', 'VAR7']

scaler = StandardScaler()
new_df[numerical_cols] = scaler.fit_transform(new_df[numerical_cols])
```

```python
new_df.head()
```

```python
new_df['Work_Life_balance'].fillna(df['Work_Life_balance'].mean(), inplace=True)
new_df['Pay_Scale'].fillna(df['Pay_Scale'].mean(), inplace=True)

```

```python
from sklearn.decomposition import PCA

# Select the relevant columns for PCA
pca_cols = ['Time_since_promotion', 'Work_Life_balance', 'Education_Level', 'Relationship_Status', 'Age']

# Apply PCA to the selected columns
pca = PCA(n_components=1)
pca_df = pca.fit_transform(new_df[pca_cols])

# Create a new column with the PCA output
new_df['Churn_Risk_Score'] = pca_df
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_df.drop('Churn_Risk_Score', axis=1), new_df['Churn_Risk_Score'], test_size=0.2, random_state=42)

```

```python
X_train = X_train.drop('Employee_ID', axis=1)
X_test = X_test.drop('Employee_ID', axis=1)

```

```python
new_df.describe()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=10, random_state=42)
# Select only numerical columns
numerical_cols = X_train.select_dtypes(exclude=['object']).columns

# Create new training and testing datasets with only numerical columns
X_train_num = X_train[numerical_cols]
X_test_num = X_test[numerical_cols]

# Train the model using the numerical columns
rf_model.fit(X_train_num, y_train)

# Make predictions using the numerical columns
y_pred = rf_model.predict(X_test_num)

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train_num, y_train)

y_pred = lr_model.predict(X_test_num)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2: {r2:.2f}")

```

```python
import matplotlib.pyplot as plt

for col in X_train_num.columns:
    plt.scatter(X_train_num[col], y_train)
    plt.xlabel(col)
    plt.ylabel('Target Variable')
    plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_pred_train = lr_model.predict(X_train_num)
mse_train = mean_squared_error(y_train, y_pred_train)
print(f"Training MSE: {mse_train:.2f}")

y_pred_test = lr_model.predict(X_test_num)
mse_test = mean_squared_error(y_test, y_pred_test)
print(f"Testing MSE: {mse_test:.2f}")

```

```python
import numpy as np

y_train_noisy = y_train + np.random.normal(0, 0.1, size=y_train.shape)
y_test_noisy = y_test + np.random.normal(0, 0.1, size=y_test.shape)

#Then, retrain the model on the noisy data:

lr_model.fit(X_train_num, y_train_noisy)

#And evaluate its performance on the noisy testing data:

y_pred_noisy = lr_model.predict(X_test_num)
mse_noisy = mean_squared_error(y_test_noisy, y_pred_noisy)
print(f"Noisy MSE: {mse_noisy:.2f}")
```


1. Trained a Random Forest model and achieved perfect scores.
2. Trained a Linear Regression model and achieved perfect scores.
3. Investigated overfitting by plotting training and testing scores.
4. Added noise to the target variable to test the model's robustness.




Note that the original data had no target variable which is churn, i use PCA AND feature engineering to create a target variable for the data.
